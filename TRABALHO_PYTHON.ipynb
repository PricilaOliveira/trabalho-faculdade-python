```python
#1 Importar os pacotes “pandas”, “matplotlib” e “seaborn”

import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
#2 Apresentando repositório de dados

url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/image/segmentation.data'
dados = pd.read_csv(url, header=2, skip_blank_lines=True)
print(dados)
```

```python
#3 Importando item 2 em um DataFrame

dados = pd.DataFrame(dados)
display(dados)
```

```python
#4 Importando os 5 primeiros registros do Dataframe do item 3

dados_df_item3 = dados.iloc[0:5]
display(dados_df_item3)
```

```python
#5 Apresentando as colunas do Dataframe do item 3

colunas_df_item3 = dados_df_item3.iloc[0:0]
display(colunas_df_item3)
```

```python
#6 Apresentando os tipos de dados das colunas do dataframe do item 3

tipos_col_df_item3 = colunas_df_item3.info()
display(tipos_col_df_item3)
```

```python
#7 Apresentando uma análise explanatória dos dados do Dataframe do item 3 utilizando a função "describe"

describe_df_item3 = dados_df_item3.describe()
display(describe_df_item3)
```

```python
#8 Excluindo dados NA (também feito no início ao utilizar "skip_blank_lines=True")

notna_df_item3 = dados_df_item3.dropna()
display(notna_df_item3)
```

```python
#9 Apresente o histograma dos dados do item 8

fig = plt.figure(figsize=(16,6))
histograma_item8 = notna_df_item3 
histograma_item8.hist()
plt.show()
```
